- Moving on to perform PCA

```python
# Performing PCA

X_var=glass_ds[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']]
pca=PCA(random_state=1)
pca.fit(X_var)
var_exp=pca.explained_variance_ratio_ # var_exp = variance in dataset; explained_variance_ratio_ = PCA method to get the ration of variance
cum_var_exp=np.cumsum(var_exp) # cum_var_exp = Cumulative Variance; cumsum() function gives the cumulative sum
var_df=pd.DataFrame(pca.explained_variance_.round(2), index=["P"+str(i) for i in range(1,10)], columns=["Explained_Variance"])
print(var_df)
plt.figure(figsize=(12,7))
plt.bar(range(1,len(cum_var_exp)+1), var_exp, align='center', label='Individual Variance', color='coral', alpha=0.8)
plt.step(range(1,len(cum_var_exp)+1), cum_var_exp, where='mid', label='Cumulative Variance', color='red')
plt.ylabel('Explained Variance Rratio')
plt.xlabel('Principal Components')
plt.xticks(np.arange(1,len(var_exp)+1,1))
plt.legend(loc='center right')
plt.show()
```

### Observations from above PCA:

- The first 4 components accounts for about 90% of the total explained variance
- We can now run the analysis again only with these four components

```python
pca_new=PCA(n_components=4)
X_reduced=pca_new.fit_transform(X_var)
```

#### _Now that we have our data Normalized, Skewness reduced and have our Feature and Target variables, let's proceed with Model development_ 

```python
# Splitting the dataset into test and training with 80% for training the model

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size =.2, random_state=25)

# Checking if the train and test have instances of all the glass types
print(np.unique(Y_train))
print(np.unique(Y_test))
```

## _Logistic Regression_

```python
accuracy=[]

for i in range(1):
    logreg=LogisticRegression(random_state=0,solver='lbfgs',multi_class='multinomial',max_iter=100)
    logreg.fit(X_train, Y_train)
    score=logreg.score(X_test,Y_test)
    accuracy.append(score)
    
print(logreg.score(X_train,Y_train))
print(score)
```

## _Decision Tree Alogrithm_

```python
accuracy=[]

for i in range(1):
    dta=DecisionTreeClassifier(random_state=0)
    dta.fit(X_train, Y_train)
    score=dta.score(X_test,Y_test)
    accuracy.append(score)

print(dta.score(X_train,Y_train))
print(score)
```

## _K-Nearest Neighbors_

```python
accuracy=[]

for i in range (2,11):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    score=knn.score(X_test,Y_test)
    accuracy.append(score)

print(knn.score(X_train,Y_train))
print(score)
```

## _SVC Classifier (Non-Linear Kernal)_

```python
accuracy=[]

for i in range(1):
    svc=SVC(gamma='auto')
    svc.fit(X_train, Y_train)
    score=svc.score(X_test,Y_test)
    accuracy.append(score)

print(svc.score(X_train,Y_train))
print(score)
```

## _SVC Classifier (Linear Kernal)_

```python
accuracy=[]

for i in range(1):
    svc=LinearSVC(random_state=0)
    svc.fit(X_train, Y_train)
    score=svc.score(X_test,Y_test)
    accuracy.append(score)

print(svc.score(X_train,Y_train))
print(score)
```

## _Random Forest Alogrithm_

```python
accuracy=[]
Range=[10,20,30,50,70,80,100,120]

for i in range(1):
    rf=RandomForestClassifier(criterion='gini', n_estimators=10, min_samples_leaf=1, min_samples_split=4, random_state=1, n_jobs=-1)
    rf.fit(X_train, Y_train)
    score=rf.score(X_test,Y_test)

print(rf.score(X_train,Y_train))
print(score)
```

```python
score=[]
Range=[10,20,30,50,70,80,100,120]

gd=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
gd.fit(X_train, Y_train)
score=gd.score(X_test,Y_test)

print(gd.score(X_train,Y_train))
print(score)
```
